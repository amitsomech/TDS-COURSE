```python
dataset_path = "../datasets/houseprices/train.csv"
dtf = pd.read_csv(dataset_path,index_col='Id')
```

```python
categorical_columns = [c for c in dtf.columns if c not in numeric_columns]
numeric_columns = dtf.dtypes[(dtf.dtypes=="float64") | (dtf.dtypes=="int64")].index.tolist()
very_numerical = [nc for nc in numeric_columns if dtf[nc].nunique()>20]
```

    -\left(q_{.5}-q_{.25}\right)}{q_{.75}-q_{.25}}$
     
$    SK_{3}=\frac{\mu-\hat{q}_{0.5}}
    {\hat{E}\left[\left|y-\hat{\mu}\right|\right]}$


$
    SK_{4}=\frac{\mu-\hat{q}_{0.5}}{\hat{\sigma}}

### Calculating for all numeric columns:

```python
stats_list=[]
for c in very_numerical:
    s1,s2,s3,s4 = sm.stats.stattools.robust_skewness(dtf[c])
    k1,k2,k3,k4 = sm.stats.stattools.robust_kurtosis(dtf[c])
    d = {"column":c,
         "s1":s1,
         "s2":s2,
         "s3":s3,
         "s4":s4,
         "k1":k1,
         "k2":k2,
         "k3":k3,
         "k4":k4,    
    }
    stats_list.append(d)
stats_df = pd.DataFrame(stats_list)
```

### Log Transform:
##### One of the most common methods to handle skewness is to apply log(x) on the data

```python
log_price = dtf.SalePrice.apply(np.log)
```

```python
log_price.hist()
```

```python
log_price.skew(),log_price.kurt()
```

##### Use KS to see how "normal" is the data now:

```python
m,s =stats._continuous_distns.norm.fit(log_price)
m,s
```

```python
sp.stats.ks_1samp(log_price.values,stats._continuous_distns.norm.cdf,args=(m,s))
```

### Square root transform:

```python
sqrt_price = dtf.SalePrice.apply(np.sqrt)
sqrt_price.hist()
```

```python
sqrt_price.skew(),sqrt_price.kurt()
```

##### Use KS to see how "normal" is the data now:

```python
m,s =stats._continuous_distns.norm.fit(sqrt_price)
sp.stats.ks_1samp(sqrt_price.values,stats._continuous_distns.norm.cdf,args=(m,s))
```

### Box-Cox Transform:

$y(\lambda)= \begin{cases}\frac{y^{\lambda}-1}{\lambda}, & \text { if } \lambda \neq 0 \\ \log y, & \text { if } \lambda=0\end{cases}$

```python
bc = sp.stats.boxcox
```

```python
bc_price, lmda  = bc(dtf.SalePrice)
```

```python
bc_price = pd.Series(bc_price)
bc_price.hist()
```

```python
lmda
```

##### Use KS to see how "normal" is the data now:

```python
m,s =stats._continuous_distns.norm.fit(bc_price)
sp.stats.ks_1samp(bc_price.values,stats._continuous_distns.norm.cdf,args=(m,s))
```
